```python
# Categorical & Continuous columns.

cont = []
cat = []
for i in df.columns : 
    if df[i].nunique() > 10 :     #Continuous 
        cont.append(i)
    else : 
        cat.append(i)             # Categorical
        
print(" Categorical : ", cat)
print(" Continous : " , cont)
```

```python
# Name, sex,age, survival status of passengers.

df[['Name', 'Sex', 'Age', 'Survived']]
```

```python
# Passengers traveling in first class.

df[df.Pclass == 1 ]
```

```python
# Average fare.

df.Fare.mean().round(1)
```

```python
# Classes of each categorical columns.

for i in cat : 
    print(i, end = "\t")
    print(df[i].unique())
    pass
```

```python
# Columns with null values.

df.isnull().sum()
```

```python
# Count of survived/non survived.  

df['Survived'].value_counts()
```

```python
# Count of male/females.

df['Sex'].value_counts()
```

```python
# Count of passengers in each class.

df['Pclass'].value_counts()
```

```python
# Count of each embarked.

df['Embarked'].value_counts()
```

```python
# Mean , median of ages of passengers. 

print("Mean   \t    : ", df['Age'].mean().round(1))
print("Median \t    : ", df['Age'].median())
```

```python
# Names of all not survived passengers.

df[df.Survived == 0 ]['Name']
```

```python
# Name of all females who survived.

df[(df.Survived == 1) & (df.Sex == 'female')]['Name']
```

```python
# Details of all males, travelling in class I who Survived.

df[(df.Survived == 1) & (df.Sex == 'male') & (df.Pclass == 1)]
```

```python
# Details of females with age >50 that survived.

df[(df.Survived == 1) & (df.Sex == 'female') & (df.Age > 50)]
```

```python
# Survival status of all passengers travelling in class I.

df[(df.Pclass == 1)]['Survived'].value_counts()
```

```python
# Percentage of survival status of all passenger travelling in class I

df[(df.Pclass == 1)]['Survived'].value_counts() / df[(df.Pclass == 1)]['Survived'].value_counts().sum() * 100 
```

```python
# Survival status of all passengers travelling in class II.

df[(df.Pclass == 2)]['Survived'].value_counts()
```

```python
# Percentage of Survival status of all passengers travelling in class II.

df[(df.Pclass == 2)]['Survived'].value_counts() /df[(df.Pclass == 2)]['Survived'].value_counts().sum() * 100
```

```python
# Survival status of all passengers travelling in class III.

df[(df.Pclass == 3)]['Survived'].value_counts()
```

```python
# Percentage of Survival status of all passengers travelling in class III.

df[(df.Pclass == 3)]['Survived'].value_counts() / df[(df.Pclass == 3)]['Survived'].value_counts().sum() * 100
```

```python
# Survival status of passengers who paid more than median fare of ticket.

df[df.Fare > df.Fare.median()]['Survived'].value_counts()
```

```python
#  Gender wise Survival status of passengers.

df.groupby('Sex')['Survived'].value_counts().unstack()
```

```python
# Passenger class, Sex wise survival status.

df.groupby(['Pclass', 'Sex'])['Survived'].value_counts().unstack()
```

```python
# Pairplot
sns.pairplot(df, hue='Survived')  # Assuming 'Survived' is the target
plt.suptitle('Pairplot of Titanic Dataset', y=1.02)
plt.show()
```

```python
# Heatmap of Correlation Matrix
plt.figure(figsize=(12,8))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Numeric Features Only)')
plt.show()
```

```python
# Histograms
df.hist(figsize=(14,10), bins=20, edgecolor='black')
plt.suptitle('Histograms of Features', y=1.02)
plt.show()
```

```python
# Boxplots
plt.figure(figsize=(10,6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Boxplot of Age vs Survived')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Boxplot of Fare vs Survived')
plt.show()
```

```python
# Scatterplots
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Scatterplot of Age vs Fare by Survival')
plt.show()
```
